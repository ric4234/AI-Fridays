### Build the `sentence embedding` pipeline using 🤗 Transformers Library

```python
import torch
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = SentenceTransformer("all-MiniLM-L6-v2")
```

More info on [all-MiniLM-L6-v2](https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2).

```python
sentence = "I go to the Technlogy Park in Pavia"
```

```python
tokenizer.tokenize(sentence)
```

```python
inputs = tokenizer(sentence)
encoded_sentence = inputs["input_ids"]
encoded_sentence
```

```python
tokenizer.decode(encoded_sentence)
```

```python
print(tokenizer.tokenize("king"))
```

```python
print(model.encode("king", convert_to_tensor=True))
```

```python
king = model.encode("king", convert_to_tensor=True)
man = model.encode("man", convert_to_tensor=True)
woman = model.encode("woman", convert_to_tensor=True)
queen = model.encode("queen", convert_to_tensor=True)
cat = model.encode("cat", convert_to_tensor=True)

```

```python
king.size()
```

```python
print(torch.dot(king,king))
print(torch.dot(man,man))
print(torch.dot(woman,woman))
print(torch.dot(queen,queen))
print("-----------------")
print(torch.dot(cat,queen))
print(torch.dot(cat,king))
```
